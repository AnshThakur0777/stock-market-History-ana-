```python
import gym
from gym import spaces

class StockTradingEnv(gym.Env):
    def __init__(self, df):
        super(StockTradingEnv, self).__init__()
        self.df = df
        self.action_space = spaces.Discrete(3)  # Actions: 0 - Hold, 1 - Buy, 2 - Sell
        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(5,))
        self.reset()

    def reset(self):
        self.current_step = 0
        self.balance = 100000  # Initial balance
        self.shares_held = 0
        self.net_worth = self.balance
        self.max_net_worth = self.balance
        self.df['Portfolio Value'] = 0
        return self._next_observation()

    def _next_observation(self):
        obs = np.array([
            self.df.loc[self.current_step, 'Open'],
            self.df.loc[self.current_step, 'High'],
            self.df.loc[self.current_step, 'Low'],
            self.df.loc[self.current_step, 'Close'],
            self.df.loc[self.current_step, 'Volume']
        ])
        return obs

    def step(self, action):
        current_price = self.df.loc[self.current_step, 'Close']
        reward = 0

        if action == 1:  # Buy
            self.shares_held += self.balance // current_price
            self.balance -= self.shares_held * current_price
        elif action == 2:  # Sell
            self.balance += self.shares_held * current_price
            self.shares_held = 0

        self.net_worth = self.balance + self.shares_held * current_price
        self.max_net_worth = max(self.max_net_worth, self.net_worth)
        reward = self.net_worth - self.balance

        self.current_step += 1
        done = self.current_step >= len(self.df) - 1

        return self._next_observation(), reward, done, {}

env = StockTradingEnv(stock_data)
```

```python
import numpy as np
import random
from collections import deque
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import gym  # Assuming we're using OpenAI Gym for the environment

# DQN Agent
class DQNAgent:
    def __init__(self, state_size, action_size, gamma=0.95, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995, learning_rate=0.001):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)  # Replay memory for experience replay
        self.gamma = gamma                # Discount factor for future rewards
        self.epsilon = epsilon            # Exploration rate for -greedy policy
        self.epsilon_min = epsilon_min    # Minimum exploration rate
        self.epsilon_decay = epsilon_decay  # Rate at which exploration decreases
        self.learning_rate = learning_rate
        self.model = self._build_model()  # Neural network for Q-value approximation

    def _build_model(self):
        # Build a simple neural network model
        model = Sequential([
            Dense(24, input_dim=self.state_size, activation="relu"),
            Dense(24, activation="relu"),
            Dense(self.action_size, activation="linear")
        ])
        model.compile(loss="mse", optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def remember(self, state, action, reward, next_state, done):
        # Store experiences in replay memory
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        # -greedy action selection
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        q_values = self.model.predict(state, verbose=0)
        return np.argmax(q_values[0])  # Action with max Q-value

    def replay(self, batch_size):
        # Experience replay to train the model
        minibatch = random.sample(self.memory, batch_size)
        states, targets = [], []

        for state, action, reward, next_state, done in minibatch:
            target = reward
            if not done:
                target += self.gamma * np.amax(self.model.predict(next_state, verbose=0)[0])

            # Get the current prediction for state and adjust only the action's Q-value
            target_f = self.model.predict(state, verbose=0)
            target_f[0][action] = target

            states.append(state[0])  # Flatten the state for fitting
            targets.append(target_f[0])  # Flatten the target for fitting

        self.model.fit(np.array(states), np.array(targets), epochs=1, verbose=0, batch_size=batch_size)

        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay  # Decrease exploration rate

# Initialize environment and DQN agent
env = gym.make("CartPole-v1")  # Substitute with any compatible Gym environment
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = DQNAgent(state_size, action_size)

# Training parameters
episodes = 10     # Number of episodes to train
batch_size = 32   # Number of samples for replay

# Training loop
for e in range(episodes):
    state = env.reset()
    state = np.reshape(state, [1, state_size])  # Reshape state for the model input

    for time in range(500):
        action = agent.act(state)  # Choose action
        next_state, reward, done, _ = env.step(action)  # Take action in the environment
        next_state = np.reshape(next_state, [1, state_size])  # Reshape next_state for the model input
        agent.remember(state, action, reward, next_state, done)  # Store experience
        state = next_state  # Move to next state

        if done:
            print(f"Episode: {e+1}/{episodes}, Score: {time}, Epsilon: {agent.epsilon:.2f}")
            break

        # Train the agent with replay if enough memories are stored
        if len(agent.memory) > batch_size:
            agent.replay(batch_size)
```

```python
# Install necessary libraries
!pip install tensorflow

# Import libraries for LSTM
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import numpy as np
import matplotlib.pyplot as plt

# Select the 'Close' price for prediction
data = stock_data['Close'].values.reshape(-1, 1)

# Normalize the data for better LSTM performance
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Prepare training data
train_data = scaled_data[:int(0.8 * len(scaled_data))]
test_data = scaled_data[int(0.8 * len(scaled_data)):]

# Create sequences of 60 days as input to predict the next day's price
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 60
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Reshape data to fit LSTM input requirements
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Build LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=1, epochs=3)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform to get actual prices
train_predict = scaler.inverse_transform(train_predict)
y_train = scaler.inverse_transform([y_train])
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform([y_test])

# Visualization
train_size = len(train_predict)
test_size = len(test_predict)

# Define train and test date ranges
train_dates = stock_data['Date'][time_step:time_step + train_size]
test_dates = stock_data['Date'][train_size + (time_step * 2) + 1:train_size + (time_step * 2) + 1 + test_size]

plt.figure(figsize=(14, 7))

# Plot actual prices
plt.plot(stock_data['Date'], data, label='Actual Close Price')

# Plot predictions for training data
plt.plot(train_dates, train_predict, label='Train Predictions')

# Plot predictions for testing data
plt.plot(test_dates, test_predict, label='Test Predictions')

# Add labels and title
plt.title('Stock Price Prediction Using LSTM')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from collections import deque
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import gym

# DQN Agent with required logging
class DQNAgent:
    def __init__(self, state_size, action_size, gamma=0.95, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995, learning_rate=0.001):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.learning_rate = learning_rate
        self.model = self._build_model()

        # Logging variables
        self.rewards_per_episode = []
        self.loss_history = []
        self.epsilon_values = []
        self.action_counts = np.zeros(action_size)  # Initialize action counts
        self.q_values = []

    def _build_model(self):
        model = Sequential([
            Dense(24, input_dim=self.state_size, activation="relu"),
            Dense(24, activation="relu"),
            Dense(self.action_size, activation="linear")
        ])
        model.compile(loss="mse", optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            action = random.randrange(self.action_size)
        else:
            q_values = self.model.predict(state, verbose=0)
            action = np.argmax(q_values[0])
            self.q_values.append(np.max(q_values))  # Record max Q-value for visualization
        self.action_counts[action] += 1  # Track actions taken
        return action

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return  # Skip replay if not enough samples

        minibatch = random.sample(self.memory, batch_size)
        states, targets = [], []

        for state, action, reward, next_state, done in minibatch:
            target = reward if done else reward + self.gamma * np.amax(self.model.predict(next_state, verbose=0)[0])
            target_f = self.model.predict(state, verbose=0)
            target_f[0][action] = target
            states.append(state[0])
            targets.append(target_f[0])

        # Train model and record loss
        history = self.model.fit(np.array(states), np.array(targets), epochs=1, verbose=0, batch_size=batch_size)
        self.loss_history.append(history.history["loss"][0])  # Record the loss

        # Decay epsilon
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# Initialize environment and agent
env = gym.make("CartPole-v1")
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = DQNAgent(state_size, action_size)

# Training parameters
episodes = 5  # Number of episodes to train
batch_size = 32

# Training loop
for e in range(episodes):
    state = env.reset()
    state = np.reshape(state, [1, state_size])
    total_reward = 0

    for time in range(500):
        action = agent.act(state)
        next_state, reward, done, _ = env.step(action)
        next_state = np.reshape(next_state, [1, state_size])
        total_reward += reward
        agent.remember(state, action, reward, next_state, done)
        state = next_state

        if done:
            agent.rewards_per_episode.append(total_reward)
            agent.epsilon_values.append(agent.epsilon)
            print(f"Episode: {e+1}/{episodes}, Score: {time}, Epsilon: {agent.epsilon:.2f}")
            break

        if len(agent.memory) > batch_size:
            agent.replay(batch_size)

# Visualization of results
plt.figure(figsize=(12, 8))

# Plot 1: Total Rewards per Episode
plt.subplot(3, 2, 1)
plt.plot(agent.rewards_per_episode)
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Total Rewards per Episode")

# Plot 2: Epsilon Decay over Episodes
plt.subplot(3, 2, 2)
plt.plot(agent.epsilon_values)
plt.xlabel("Episode")
plt.ylabel("Epsilon")
plt.title("Epsilon Decay over Episodes")

# Plot 3: Loss over Training Steps
plt.subplot(3, 2, 3)
plt.plot(agent.loss_history)
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("Training Loss over Time")

# Plot 4: Smoothed Reward Curve (Moving Average of Rewards)
window = 10
smoothed_rewards = np.convolve(agent.rewards_per_episode, np.ones(window) / window, mode='valid')
plt.subplot(3, 2, 4)
plt.plot(smoothed_rewards)
plt.xlabel("Episode")
plt.ylabel("Smoothed Reward")
plt.title("Smoothed Reward Curve (Moving Average)")

# Plot 5: Action Distribution
plt.subplot(3, 2, 5)
plt.bar(range(action_size), agent.action_counts)
plt.xlabel("Action")
plt.ylabel("Frequency")
plt.title("Distribution of Actions Taken")

# Plot 6: Distribution of Q-Values
plt.subplot(3, 2, 6)
plt.hist(agent.q_values, bins=20)
plt.xlabel("Q-Value")
plt.ylabel("Frequency")
plt.title("Distribution of Q-Values")

plt.tight_layout()
plt.show()
```
